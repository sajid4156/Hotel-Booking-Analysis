##**Step 1: To start with basics of data cleaning,find out the columns that have any Null or missing values and replace those with integers or strings.**


```python
bookings_raw_df.isnull().sum()
```

```python
bookings_raw_df['children'].fillna(0,inplace=True)
bookings_raw_df['country'].fillna('OTHERS',inplace=True)
bookings_raw_df['children'] = bookings_raw_df['children'].astype(int)
```

```python
bookings_raw_df.info()
```

##**Step 2: To remove columns which are not required for analysis and clubbing others**


```python
bookings_raw_df.columns
```

```python
selected_columns= ['hotel', 'is_canceled', 'lead_time', 'arrival_date_year',
       'arrival_date_month', 'arrival_date_week_number',
       'arrival_date_day_of_month', 'stays_in_weekend_nights',
       'stays_in_week_nights', 'adults', 'children', 'meal',
       'country', 'market_segment', 'distribution_channel',
       'is_repeated_guest', 'previous_cancellations',
       'previous_bookings_not_canceled', 'reserved_room_type',
       'assigned_room_type', 'deposit_type', 'customer_type', 'adr',
       'required_car_parking_spaces', 'total_of_special_requests']
```

```python
# Creating a copy so as not to disturb original dataset.
bookings_df = bookings_raw_df[selected_columns].copy()
```

```python
# dropping babies column as they are not accounted into monetary aspects.
bookings_df['Total_Guests'] = bookings_df['adults'] + bookings_df['children']
```

```python
bookings_df = bookings_df.drop(['adults', 'children'], axis=1)
```

```python
# removing rows which contain total guests as Zero.
bookings_df= bookings_df[bookings_df.Total_Guests > 0]
```

```python
bookings_df. head(10)
```

##**Step 3: To seperate and create different datasets of both the hotel types.**

```python
bookings_df.hotel.unique()
```

```python
# selecting only those rows which donot have booking cancellations.
df= bookings_df[bookings_df.is_canceled == 0]
df.sample()
```

```python
resorthotel_df = df [df.hotel == "Resort Hotel"]
cityhotel_df= df [df.hotel == "City Hotel"]
```

##**5. Visualizing monthly bookings and monthly cancellations**

```python
month_sorted = ['January','February','March','April','May','June','July','August','September','October','November','December']
plt.title("Monthly bookings")
sns.barplot(x= 'arrival_date_month', y= 'no_of_bookings', hue= df_copy['hotel'], estimator= sum, data= df_copy)
plt.xticks(rotation = 90)
plt.show()
```

```python
df_copy1= bookings_df.copy()
df_copy1['no_of_bookings']= 1
df_copy1.loc[(df_copy1.arrival_date_month== 'July')| (df_copy1.arrival_date_month== 'August'),'no_of_bookings']/=3
df_copy1.loc[~((df_copy1.arrival_date_month== 'July')| (df_copy1.arrival_date_month== 'August')),'no_of_bookings']/=2
df_copy1.sample(100)
b= df_copy1[df_copy1.is_canceled == 1]
month_sorted = ['January','February','March','April','May','June','July','August','September','October','November','December']
sns.set(style="dark")
plt.title("Monthly cancellations")
sns.barplot(x= 'arrival_date_month', y= 'no_of_bookings', hue= df_copy1['hotel'], estimator= sum, data= df_copy1)
plt.xticks(rotation = 90)
plt.show()
```

**Conclusion:It is observed that in case of City Hotel, months with high bookings (May, June, September, October) also witnessed more cancellations. #**Q1: What is the relationship between lead time and cancellation?**

#**Q2: What is the relationship between deposit type and cancellation?**


```python
bookings_df[bookings_df.deposit_type== 'Refundable'].deposit_type.count()
```

```python
plt.title("Booking Canceled or not by Deposit type")
sns.countplot(x='deposit_type',data=bookings_df ,hue='is_canceled')
```

##***Answer: Around 25% of bookings were cancelled by guests with no deposit. These numbers are huge if the hotels were not able to replace the cancelled bookings in time. So it's obvious that guests who do not pay any deposit while booking are likely to cancel more reservations. Also it is interesting to note that non-refundable deposits had more cancellation than refundable deposits. Logically one would have assumed that refundable deposits have more cancellation as hotel rates are usually higher for refundable deposit type rooms and customers pay more in anticipation of cancellation.***##

#**Q3: What is the relationship between market segment and cancellation?**

#**Q5: How much does the price in hotels vary over the year?**


```python
df['adr_pp']= df.adr/df.Total_Guests
monthlyadr_df= df.groupby (['hotel','arrival_date_month'])['adr_pp'].mean()
monthlyadr_df=pd.DataFrame({"hotel":(map(lambda x:x[0],monthlyadr_df.index)), "month": list(map(lambda x:x[1],monthlyadr_df.index)), "mean": monthlyadr_df.values})
monthlyadr_df
```

```python
month_sorted = ['January','February','March','April','May','June','July','August','September','October','November','December']
sns.barplot (y= 'mean', x= 'month', data= monthlyadr_df, hue= 'hotel', order= month_sorted)
plt.xticks(rotation = 90)
plt.show()
```

```python
# plotting variation in adr with room type
sns.barplot(data = df, x = 'reserved_room_type', y = 'adr_pp', hue = 'hotel')
```

```python
sns.barplot(x= 'market_segment', y= 'adr_pp', hue= 'reserved_room_type', data= df)
```

##***Answer: This plot clearly shows that prices in the Resort Hotel are much higher during the summer and prices of city hotel varies less and is most expensive during Spring and Autumn and these vary with room type and meal type.***##
